```python
import torch
from diffusers import StableDiffusionInpaintPipeline
from PIL import Image
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab import files

def auto_mask_bulb(image_pil, threshold=220):
    """Auto-detects bulb/light areas"""
    image = np.array(image_pil.convert("RGB"))
    hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
    _, mask_v = cv2.threshold(hsv[:,:,2], threshold, 255, cv2.THRESH_BINARY)
    _, mask_s = cv2.threshold(hsv[:,:,1], 40, 255, cv2.THRESH_BINARY_INV)
    combined_mask = cv2.bitwise_and(mask_v, mask_s)
    kernel = np.ones((7,7), np.uint8)
    return cv2.morphologyEx(combined_mask, cv2.MORPH_CLOSE, kernel, iterations=3)

def ai_remove_bulb(image_pil, prompt="a clean ceiling with no bulb or light", strength=0.85, steps=25):
    """Performs AI inpainting"""
    mask = auto_mask_bulb(image_pil)
    pipe = StableDiffusionInpaintPipeline.from_pretrained(
        "stabilityai/stable-diffusion-2-inpainting",
        torch_dtype=torch.float16,
    ).to("cuda")
    return pipe(
        prompt=prompt,
        image=image_pil,
        mask_image=Image.fromarray(mask),
        strength=strength,
        num_inference_steps=steps,
    ).images[0]

# Upload image
uploaded = files.upload()
image_path = list(uploaded.keys())[0]
original_image = Image.open(image_path)

# Process image
result = ai_remove_bulb(original_image)

# Show results
plt.figure(figsize=(12,6))
plt.subplot(1,2,1); plt.imshow(original_image); plt.title("Original")
plt.subplot(1,2,2); plt.imshow(result); plt.title("Processed")
plt.show()

# Save result
result.save("cleaned_image.jpg")
files.download("cleaned_image.jpg")
```
